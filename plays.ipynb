```python
import os
import glob
from pydub import AudioSegment

video_dir = r'e:\videos\cuts'  # Path where the videos are located
extension_list = ('*.mp4', '*.m4v')

os.chdir(video_dir)
for extension in extension_list:
    print(f'scanning for {extension} ...')
    for video in glob.glob(extension):
        video = os.path.join(r'e:\videos\cuts', video)
        print(f'video={video}')

        mp3_filename = os.path.splitext(os.path.basename(video))[0] + '.mp3'
        mp3_filename = os.path.join(r'h:\temp', mp3_filename)
        print(f'mp3_filename = {mp3_filename}')
        AudioSegment.from_file(video).export(mp3_filename, format='mp3')
```

```python
from glob import glob
from pydub import AudioSegment
import random

n = 0
for mp3_file in glob(recursive=True, pathname=r"E:\music\vkmusic\*.mp3"):
    print(f'loading #{n:02d} {mp3_file}')
    playlist_songs.append(AudioSegment.from_mp3(mp3_file))
    
    n += 1
    if n > 50:
        break
    
first_song = playlist_songs.pop(0)

# let's just include the first 30 seconds of the first song (slicing
# is done by milliseconds)
print(f'cut first ...')
beginning_of_song = first_song[:11 * 100]

playlist = beginning_of_song
for song in playlist_songs:
    print(f'applying cut ...')
    song = song[:5 * 1000]
    print(f'adding {song.duration_seconds:.0f} seconds of {song} [frames={song.frame_count()}]')
    # We don't want an abrupt stop at the end, so let's do a 10 second crossfades

    playlist = playlist.append(song, crossfade=(5 * 100))

    ss = playlist_songs[random.randrange(0, len(playlist_songs) - 1)]
    title = ss
    s = ss[:10 * 1000]
    print(f'adding#[{title}] {s.duration_seconds:.0f} seconds of {s} [frames={s.frame_count()}]')

    playlist = playlist.append(s, crossfade=(5 * 100))

    if random.randrange(0, len(playlist_songs)) == 9:
        break

# let's fade out the end of the last song
playlist = playlist.fade_out(30)

# hmm I wonder how long it is... ( len(audio_segment) returns milliseconds )
playlist_length = len(playlist) / (100 * 10)
aa = rf"h:\temp\{playlist_length:000000000.0f}_minute_playlist.mp3"
# lets save it!
with open(aa, 'wb') as out_f:
    print(f'processing {aa}')
    playlist.export(out_f, format='mp3')

print(f'done')
```
